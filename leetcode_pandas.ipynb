```python
data = [[9, 'Hercy', None, 43], [6, 'Alice', 9, 41], [4, 'Bob', 9, 36], [2, 'Winston', None, 37]]
employees = pd.DataFrame(data, columns=['employee_id', 'name', 'reports_to', 'age']).astype({'employee_id':'Int64', 'name':'object', 'reports_to':'Int64', 'age':'Int64'})
```

```python
import numpy as np
import pandas as pd
```

```python
merged=pd.merge(employees,employees , left_on='employee_id' ,right_on='reports_to' ,how='inner' , suffixes=('_employee', '_manager'))
result = merged.groupby(['employee_id_employee','name_employee']).agg(
    reports_count=('reports_to_manager','count'),
    average_age=('age_manager','mean')
).reset_index()

result['average_age'] = np.ceil(result['average_age'])
#result['average_age'] = result['average_age'].round(0)
result = result.rename(columns={
    'employee_id_employee': 'employee_id',
    'name_employee': 'name',
    'reports_count': 'reports_count',
    'average_age': 'average_age'
})
sorted_result=result.sort_values(by='employee_id',ascending=True)
sorted_result

```

```python
import pandas as pd

def second_highest_salary(employee: pd.DataFrame) -> pd.DataFrame:

    df = employee.drop_duplicates(['salary'])

    if len(df) == 1:
        return pd.DataFrame([{'SecondHighestSalary':None}])

    df = (df.sort_values(['salary'])
            .tail(2).head(1)
            .rename(columns = {'salary':'SecondHighestSalary'}))

    return df[['SecondHighestSalary']]
```
